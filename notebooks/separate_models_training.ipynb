# 🌾 Separate Models Training: Crops & Fruits

**Purpose**: Train specialized models for crops and fruits separately

**Benefits**: Higher accuracy, no cross-category confusion, better user experience

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
import pickle
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
```

## Step 1: Load and Split Dataset

**Separate crops and fruits for specialized training**

```python
# Upload your dataset
from google.colab import files
print("📤 Please upload your dataset (crops + fruits)")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]
    df = pd.read_csv(filename)
    
    print(f"📊 Dataset Shape: {df.shape}")
    print(f"🌾 Unique Items: {len(df['label'].unique())}")
    print(f"📝 All Labels: {sorted(df['label'].unique())}")
else:
    print("❌ No file uploaded")
    exit()
```

```python
# Define crop and fruit categories
CROPS = ['rice', 'maize', 'chickpea', 'kidneybeans', 'pigeonpeas', 
         'mothbeans', 'mungbean', 'blackgram', 'lentil', 'cotton', 'jute', 'coffee']

FRUITS = ['pomegranate', 'banana', 'mango', 'grapes', 'watermelon', 
          'muskmelon', 'apple', 'orange', 'papaya', 'coconut']

print(f"🌾 Crops ({len(CROPS)}): {CROPS}")
print(f"🍎 Fruits ({len(FRUITS)}): {FRUITS}")

# Split dataset
crop_data = df[df['label'].isin(CROPS)].copy()
fruit_data = df[df['label'].isin(FRUITS)].copy()

print(f"\n📊 Crop Dataset: {crop_data.shape}")
print(f"🍎 Fruit Dataset: {fruit_data.shape}")
print(f"🌾 Crop Types: {sorted(crop_data['label'].unique())}")
print(f"🍎 Fruit Types: {sorted(fruit_data['label'].unique())}")
```

## Step 2: Train Crop Model

```python
# Prepare crop data
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
target = 'label'

X_crop = crop_data[features]
y_crop = crop_data[target]

# Split crop data
X_crop_train, X_crop_test, y_crop_train, y_crop_test = train_test_split(
    X_crop, y_crop, test_size=0.2, random_state=42, stratify=y_crop
)

print(f"🌾 Crop Training Set: {X_crop_train.shape}")
print(f"🌾 Crop Testing Set: {X_crop_test.shape}")
print(f"🌾 Crop Types: {len(y_crop_train.unique())}")

# Train crop model
crop_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=15,
    min_samples_split=5,
    n_jobs=-1
)

print("🚀 Training Crop Model...")
crop_model.fit(X_crop_train, y_crop_train)
print("✅ Crop Model Training Completed!")

# Evaluate crop model
y_crop_pred = crop_model.predict(X_crop_test)
crop_accuracy = accuracy_score(y_crop_test, y_crop_pred)
print(f"🎯 Crop Model Accuracy: {crop_accuracy:.4f} ({crop_accuracy*100:.2f}%)")
```

## Step 3: Train Fruit Model

```python
# Prepare fruit data
X_fruit = fruit_data[features]
y_fruit = fruit_data[target]

# Split fruit data
X_fruit_train, X_fruit_test, y_fruit_train, y_fruit_test = train_test_split(
    X_fruit, y_fruit, test_size=0.2, random_state=42, stratify=y_fruit
)

print(f"🍎 Fruit Training Set: {X_fruit_train.shape}")
print(f"🍎 Fruit Testing Set: {X_fruit_test.shape}")
print(f"🍎 Fruit Types: {len(y_fruit_train.unique())}")

# Train fruit model
fruit_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=15,
    min_samples_split=5,
    n_jobs=-1
)

print("🚀 Training Fruit Model...")
fruit_model.fit(X_fruit_train, y_fruit_train)
print("✅ Fruit Model Training Completed!")

# Evaluate fruit model
y_fruit_pred = fruit_model.predict(X_fruit_test)
fruit_accuracy = accuracy_score(y_fruit_test, y_fruit_pred)
print(f"🍎 Fruit Model Accuracy: {fruit_accuracy:.4f} ({fruit_accuracy*100:.2f}%)")
```

## Step 4: Save Both Models

```python
# Save crop model
with open('crop_model_specialized.pkl', 'wb') as file:
    pickle.dump(crop_model, file)

with open('crop_features_specialized.pkl', 'wb') as file:
    pickle.dump(features, file)

print("💾 Crop Model Saved: crop_model_specialized.pkl")
print("💾 Crop Features Saved: crop_features_specialized.pkl")

# Save fruit model
with open('fruit_model_specialized.pkl', 'wb') as file:
    pickle.dump(fruit_model, file)

with open('fruit_features_specialized.pkl', 'wb') as file:
    pickle.dump(features, file)

print("💾 Fruit Model Saved: fruit_model_specialized.pkl")
print("💾 Fruit Features Saved: fruit_features_specialized.pkl")

# Test both models
print("\n🧪 Testing Specialized Models:")

# Test crop model
test_crop_input = [[90, 42, 43, 20.8, 82.0, 6.5, 202.9]]  # Rice conditions
crop_pred = crop_model.predict(test_crop_input)[0]
crop_conf = crop_model.predict_proba(test_crop_input).max() * 100
print(f"🌾 Crop Model Test: {crop_pred} ({crop_conf:.2f}% confidence)")

# Test fruit model
test_fruit_input = [[95, 75, 85, 35.0, 95.0, 7.8, 250.0]]  # Papaya conditions
fruit_pred = fruit_model.predict(test_fruit_input)[0]
fruit_conf = fruit_model.predict_proba(test_fruit_input).max() * 100
print(f"🍎 Fruit Model Test: {fruit_pred} ({fruit_conf:.2f}% confidence)")

print("\n📁 Download these files for your app:")
print("1. crop_model_specialized.pkl")
print("2. crop_features_specialized.pkl")
print("3. fruit_model_specialized.pkl")
print("4. fruit_features_specialized.pkl")
```
