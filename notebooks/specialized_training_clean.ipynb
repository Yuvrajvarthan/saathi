```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pickle
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
```

## Step 1: Upload and Load Separated Datasets

```python
# Upload your separated datasets
from google.colab import files
print("📤 Please upload both separated datasets:")
print("  1. crops_dataset.csv")
print("  2. fruits_dataset.csv")
uploaded = files.upload()

if len(uploaded) == 2:
    # Load both datasets
    crop_df = pd.read_csv('crops_dataset.csv')
    fruit_df = pd.read_csv('fruits_dataset.csv')
    
    print(f"🌾 Crop Dataset Shape: {crop_df.shape}")
    print(f"🍎 Fruit Dataset Shape: {fruit_df.shape}")
    
    print(f"🌾 Crop Types: {sorted(crop_df['label'].unique())}")
    print(f"🍎 Fruit Types: {sorted(fruit_df['label'].unique())}")
else:
    print("❌ Please upload both datasets!")
    exit()
```

## Step 2: Train Crop Model

```python
# Prepare crop data
features = ['N', 'P', 'K', 'temperature', 'humidity', 'ph', 'rainfall']
target = 'label'

X_crop = crop_df[features]
y_crop = crop_df[target]

print(f"🌾 Crop Features Shape: {X_crop.shape}")
print(f"🌾 Crop Target Shape: {y_crop.shape}")
print(f"🌾 Unique Crops: {len(y_crop.unique())}")
print(f"🌾 Crop Types: {sorted(y_crop.unique())}")

# Split crop data with stratification
X_crop_train, X_crop_test, y_crop_train, y_crop_test = train_test_split(
    X_crop, y_crop, test_size=0.2, random_state=42, stratify=y_crop
)

print(f"\n🌾 Crop Training Set: {X_crop_train.shape}")
print(f"🌾 Crop Testing Set: {X_crop_test.shape}")

# Train optimized crop model
crop_model = RandomForestClassifier(
    n_estimators=300,           # More trees for better accuracy
    random_state=42,
    max_depth=20,              # Deeper trees
    min_samples_split=4,        # Better splits
    min_samples_leaf=2,         # Prevent overfitting
    max_features='sqrt',        # Optimal feature selection
    n_jobs=-1                 # Use all CPU cores
)

print("🚀 Training Crop Model...")
crop_model.fit(X_crop_train, y_crop_train)
print("✅ Crop Model Training Completed!")

# Evaluate crop model
y_crop_pred = crop_model.predict(X_crop_test)
crop_accuracy = accuracy_score(y_crop_test, y_crop_pred)
print(f"\n🎯 Crop Model Accuracy: {crop_accuracy:.4f} ({crop_accuracy*100:.2f}%)")

# Get confidence scores
crop_confidences = crop_model.predict_proba(X_crop_test).max(axis=1)
avg_crop_confidence = np.mean(crop_confidences) * 100
print(f"📊 Average Crop Confidence: {avg_crop_confidence:.2f}%")

# Detailed classification report
print(f"\n📊 Crop Classification Report:")
print(classification_report(y_crop_test, y_crop_pred))
```

## Step 3: Train Fruit Model

```python
# Prepare fruit data
X_fruit = fruit_df[features]
y_fruit = fruit_df[target]

print(f"🍎 Fruit Features Shape: {X_fruit.shape}")
print(f"🍎 Fruit Target Shape: {y_fruit.shape}")
print(f"🍎 Unique Fruits: {len(y_fruit.unique())}")
print(f"🍎 Fruit Types: {sorted(y_fruit.unique())}")

# Split fruit data with stratification
X_fruit_train, X_fruit_test, y_fruit_train, y_fruit_test = train_test_split(
    X_fruit, y_fruit, test_size=0.2, random_state=42, stratify=y_fruit
)

print(f"\n🍎 Fruit Training Set: {X_fruit_train.shape}")
print(f"🍎 Fruit Testing Set: {X_fruit_test.shape}")

# Train optimized fruit model
fruit_model = RandomForestClassifier(
    n_estimators=300,           # More trees for better accuracy
    random_state=42,
    max_depth=20,              # Deeper trees
    min_samples_split=4,        # Better splits
    min_samples_leaf=2,         # Prevent overfitting
    max_features='sqrt',        # Optimal feature selection
    n_jobs=-1                 # Use all CPU cores
)

print("🚀 Training Fruit Model...")
fruit_model.fit(X_fruit_train, y_fruit_train)
print("✅ Fruit Model Training Completed!")

# Evaluate fruit model
y_fruit_pred = fruit_model.predict(X_fruit_test)
fruit_accuracy = accuracy_score(y_fruit_test, y_fruit_pred)
print(f"\n🎯 Fruit Model Accuracy: {fruit_accuracy:.4f} ({fruit_accuracy*100:.2f}%)")

# Get confidence scores
fruit_confidences = fruit_model.predict_proba(X_fruit_test).max(axis=1)
avg_fruit_confidence = np.mean(fruit_confidences) * 100
print(f"📊 Average Fruit Confidence: {avg_fruit_confidence:.2f}%")

# Detailed classification report
print(f"\n📊 Fruit Classification Report:")
print(classification_report(y_fruit_test, y_fruit_pred))
```

## Step 4: Save Clean Specialized Models

```python
# Save crop model
with open('crop_model_clean.pkl', 'wb') as file:
    pickle.dump(crop_model, file)

with open('crop_features_clean.pkl', 'wb') as file:
    pickle.dump(features, file)

print("💾 Crop Model Saved: crop_model_clean.pkl")
print("💾 Crop Features Saved: crop_features_clean.pkl")

# Save fruit model
with open('fruit_model_clean.pkl', 'wb') as file:
    pickle.dump(fruit_model, file)

with open('fruit_features_clean.pkl', 'wb') as file:
    pickle.dump(features, file)

print("💾 Fruit Model Saved: fruit_model_clean.pkl")
print("💾 Fruit Features Saved: fruit_features_clean.pkl")
```

## Step 5: Test Clean Models

```python
# Test crop model with clear crop conditions
test_crop_inputs = [
    [90, 42, 43, 20.8, 82.0, 6.5, 202.9],  # Rice
    [50, 50, 50, 25.0, 70.0, 7.0, 150.0],  # Maize
    [117, 46, 19, 23.9, 79.8, 6.9, 80.3]   # Cotton
]

print("🧪 Testing Crop Model:")
for i, test_input in enumerate(test_crop_inputs):
    crop_pred = crop_model.predict([test_input])[0]
    crop_conf = crop_model.predict_proba([test_input]).max() * 100
    print(f"  Test {i+1}: {crop_pred} ({crop_conf:.2f}% confidence)")

# Test fruit model with clear fruit conditions
test_fruit_inputs = [
    [95, 75, 85, 35.0, 95.0, 7.8, 250.0],  # Papaya
    [85, 65, 75, 30.0, 80.0, 6.8, 180.0],  # Mango
    [70, 50, 60, 25.0, 75.0, 7.0, 150.0]   # Grapes
]

print("\n🧪 Testing Fruit Model:")
for i, test_input in enumerate(test_fruit_inputs):
    fruit_pred = fruit_model.predict([test_input])[0]
    fruit_conf = fruit_model.predict_proba([test_input]).max() * 100
    print(f"  Test {i+1}: {fruit_pred} ({fruit_conf:.2f}% confidence)")

print(f"\n📊 Model Performance Summary:")
print(f"🌾 Crop Model: {crop_accuracy*100:.2f}% accuracy, {avg_crop_confidence:.2f}% avg confidence")
print(f"🍎 Fruit Model: {fruit_accuracy*100:.2f}% accuracy, {avg_fruit_confidence:.2f}% avg confidence")
```
